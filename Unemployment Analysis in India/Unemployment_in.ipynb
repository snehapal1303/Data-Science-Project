Heat Map

```python
plt.figure(figsize=(10,7))
plt.style.use('seaborn-whitegrid')
sns.heatmap(df.corr(), annot = True, cmap = "Oranges_r")
plt.show()
```

```python
df.columns=["States", "Date", "Frequency", "Estimated Unemployment Rate (%)", "Estimated Employed", "Estimated Labour Participation Rate (%)","Region","longitude","latitude"]
plt.title("Unemployment in India")
sns.histplot(x="Estimated Employed", hue="Region", data=df)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title("Unemployment in India")
sns.histplot(x="Estimated Unemployment Rate (%)", hue="Region", data=df)
plt.show()
```

```python
# Creating line chart
plt.plot(df['Estimated Employed'], label=' Estimated Employed')
plt.plot(df['Estimated Unemployment Rate (%)'], label=' Estimated Unemployment Rate (%)')
plt.plot(df['Estimated Labour Participation Rate (%)'], label=' Estimated Labour Participation Rate (%)')
plt.plot(df['Region'], label='Region')

# Adding labels and titles
plt.xlabel('Sample')
plt.ylabel('Measurement')
plt.title('Line Chart - Iris Dataset')

# Adding Legend
plt.legend()

# Displaying the line chart
plt.show()
```

```python
fig = px.scatter(df, x="Estimated Labour Participation Rate (%)", y="Estimated Unemployment Rate (%)", color="Region")
```

```python
fig.show()
```

```python
unemployment = df[["States","Region","Estimated Unemployment Rate (%)"]]
figure = px.sunburst(unemployment, path=["Region","States"],values="Estimated Unemployment Rate (%)",width=650, height=650, color_continuous_scale="RdY1Gn",title="Unemployment Rate Of India")
figure.show()
```

Bar Graph

```python
figg = px.bar(df,x="Region",y="Estimated Unemployment Rate (%)", color="Region", title="Bar Graph Of Unemployment", template='plotly')
figg.update_layout(xaxis={'categoryorder':'total descending'})
figg.show()
```

Box Plot

```python
figg = px.box(df,x="Region",y="Estimated Unemployment Rate (%)", color="Region", title="Box Plot Of Unemployment", template='plotly')
figg.update_layout(xaxis={'categoryorder':'total descending'})
figg.show()
```
